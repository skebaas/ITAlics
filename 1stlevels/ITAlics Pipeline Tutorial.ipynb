ITAlics Pipeline Tutorial:

This jupyter notebook will show the user how to run participants through the Pipeline up through and including first levels

The pipeline utilzes anaconda environments to handle dependencies and set environmental variables.  If you started a new terminal session, you must activate the ITAlics pipeline conda environment

skebaas@Peridot:~$ `conda activate /usr/local/software/ITAlics/conda`

Now that your paths and environmental variables are set, you can easily run fmriprep.  There is a custom bash script that already has the necessarry flags. A help flag can be displayed using `fmriprep -h`

Here is an example of how to run a subject:

(/usr/local/software/ITAlics/conda)skebaas@Peridot:/data/D2/data$ `fmriprep -s /data/D2/data/Nifti/BPD/sub-50003 -o /data/D2/data/Preprocessed/BPD -w /data/D2/data/work`

After fmriprep has finished running you will have to move the BHV folder.
You will need to move it into the Preprocessed folder as seen in the following image

It is now time to run the first levels!
Since you've already activated your conda environment and moved the BHV folder, this step is simple!

I will show you an example using the a pre-processed subject in /data/D2/data/Preprocessed_Henry/fmriprep

(/usr/local/software/ITAlics/conda)skebaas@Peridot:/data/D2/data/Preprocessed_Henry/fmriprep$ `1stlevel_analysis.py /data/D2/data/Preprocessed_Henry/fmriprep/sub-50148`

You may notice in the attached image that an absolute path for the subject is not neccessary for the 1st levels script.  Use whatever you are more comfortable with! The 1st levels should take ~1 hour to run to completion.  

You can also run individual task by specifying one of the following flags:
`1stlevel_analysis.py -reward sub-50148`

`1stlevel_analysis.py -efnback sub-50148`

`1stlevel_analysis.py -dynamic_faces sub-50148`
